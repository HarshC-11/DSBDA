```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Cleavland.csv')
df2 = pd.read_csv('hung.csv')
df3 = pd.read_csv('Switzerland.csv')
```

```python
merged_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
merged_df = merged_df.reset_index(drop=True)
```

```python
merged_df.head()
```

```python
merged_df.replace('?', np.nan, inplace=True)
```

```python
merged_df.dropna(inplace=True)
```

```python
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
# Plot histograms for selected variables
variables = ['age', 'chol', 'thalach']
merged_df[variables].hist(bins=10, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
# Plot bar plots for selected variables
variables = ['cp', 'restecg', 'slope']
for var in variables:
    plt.figure(figsize=(8, 6))
    merged_df[var].value_counts().plot(kind='bar')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.show()
```

```python
# Plot line chart for selected variables
variables = ['age', 'chol', 'thalach']
for var in variables:
    plt.figure(figsize=(10, 6))
    plt.plot(merged_df[var])
    plt.xlabel('Index')
    plt.ylabel(var)
    plt.show()
```

```python
# Plot scatter plot for selected variables
sns.scatterplot(x='chol', y='thalach', data=merged_df, hue='num')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.show()

```
