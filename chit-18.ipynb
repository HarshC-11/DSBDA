```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Heart Disease dataset into a DataFrame
df = pd.read_csv('merged_dataset.csv')
```

```python
# Plot histograms for selected variables
variables = ['age', 'chol', 'thalach']
for var in variables:
    plt.figure(figsize=(8, 6))
    plt.hist(df[var], bins=10, edgecolor='black')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.title('Histogram: {}'.format(var))
    plt.show()
```

```python
# Plot pie charts for selected variables
variables = ['sex', 'cp', 'num']
for var in variables:
    plt.figure(figsize=(6, 6))
    df[var].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.ylabel('')
    plt.title('Pie Chart: {}'.format(var))
    plt.show()
```

```python
# Plot box plots for selected variables
variables = ['age', 'chol', 'thalach']
for var in variables:
    plt.figure(figsize=(6, 6))
    plt.boxplot(df[var])
    plt.ylabel(var)
    plt.title('Box Plot: {}'.format(var))
    plt.show()
```

```python
# Plot scatter plots for selected variables
variables = ['age', 'chol', 'thalach']
for var in variables:
    plt.figure(figsize=(8, 6))
    plt.scatter(df[var], df['num'])
    plt.xlabel(var)
    plt.ylabel('Target')
    plt.title('Scatter Plot: {} vs Target'.format(var))
    plt.show()
```

```python
# Plot bar chart for selected variables
variables = ['sex', 'cp', 'num']
for var in variables:
    plt.figure(figsize=(6, 6))
    df[var].value_counts().plot(kind='bar')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.title('Bar Chart: {}'.format(var))
    plt.show()
```
